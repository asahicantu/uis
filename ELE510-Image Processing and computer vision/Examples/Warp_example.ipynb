```python
# Import the packages that are useful inside the definition of the weakPerspective function
import math
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
```

```python
img = cv2.imread('stinkbug.png')
Igray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(Igray, cmap='gray')
```

```python
rows,cols = Igray.shape
M = np.float32([[1,0,100],[0,1,50]])
M2 = np.float32([[0, -1,300],[1, 0, 0]])
print(M)
print(M2)
dst = cv2.warpAffine(Igray,M,(cols,rows))
dst2 = cv2.warpAffine(Igray,M2,(cols,rows))
plt.subplot(311),plt.imshow(Igray,cmap='gray'),plt.title('Input')
plt.subplot(312),plt.imshow(dst,cmap='gray'),plt.title('Output')
plt.subplot(313),plt.imshow(dst2,cmap='gray'),plt.title('Output')

plt.show()
```

```python
kernel = np.ones((5,5),np.float32)/25
print(kernel)
dst = cv2.filter2D(Igray,-1,kernel)
print(dst)
plt.subplot(121),plt.imshow(Igray,cmap='gray'),plt.title('Input')
plt.subplot(122),plt.imshow(dst,cmap='gray'),plt.title('Output')
plt.show()
```
